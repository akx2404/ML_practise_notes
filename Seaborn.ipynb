<h3> Correlation with Heatmap </h3>

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
iris.corr()
```

```python
ax=sns.heatmap(iris.corr())
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

<h3> JoinPlot </h3>
    <br>A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='sepal_width',y='petal_width',data=iris,kind='hex')
```

```python
sns.jointplot(x='sepal_width',y='petal_width',data=iris,kind='reg')
```

<h3> PairPlot </h3>
    <br>A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

<h3> DistPlot </h3>
    <br>Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(iris['petal_width'])
```

```python
sns.distplot(iris['petal_width'],kde=False,bins=10)
```

<h3> Count Plot </h3>

```python
sns.countplot('species',data=iris)
```

```python
sns.countplot(y='species',data=iris)
```

<h3> Bar Plot </h3>

```python
sns.barplot(x='species',y='petal_width',data=iris)
```

<h3> Box Plot </h3>
    <br>A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('species','petal_width', data=iris, orient='v')
```

<h3> Violin plot </h3>
    <br>Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="petal_width", y="species", data=iris)
```

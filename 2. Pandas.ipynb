<b> Reading different data sources with pandas </b>

```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'
            'x,y,1\n'
            'a,b,2\n'
            'c,d,3')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
## Reading from specific columns
## the lambda function given in the code(refer to page 17 of codiary)

df=pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in ['COL1',
                                                               'COL3'])
```

```python
df.to_csv('Test.csv')
```

```python
data2 = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11')
```

```python
print(data2)
```

```python
df=pd.read_csv(StringIO(data2), dtype=object)
```

```python
df['a'][1]
```
